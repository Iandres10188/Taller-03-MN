## Polinomios de Lagrange

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import lagrange

# Conjuntos de datos
lagrange_data = [
    [(0, 0), (30, 0.5), (60, np.sqrt(3)/2), (90, 1)],
    [(1, 1), (2, 2), (3, 2)],
    [(-2, 5), (1, 7), (3, 11), (7, 34)]
]

# Interpolacin y grfica
for i, points in enumerate(lagrange_data, start=1):
    x_vals_data = np.array([pt[0] for pt in points])
    y_vals_data = np.array([pt[1] for pt in points])
    
    poly = lagrange(x_vals_data, y_vals_data)
    
    x_interp = np.linspace(min(x_vals_data) - 1, max(x_vals_data) + 1, 400)
    y_interp = poly(x_interp)
    
    plt.figure(figsize=(8, 5))
    plt.plot(x_interp, y_interp, label='Polinomio de Lagrange')
    plt.scatter(x_vals_data, y_vals_data, color='red', zorder=5)
    plt.title(f'Interpolacin de Lagrange - Conjunto {i}')
    plt.grid(True)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.tight_layout()
    plt.show()
```
